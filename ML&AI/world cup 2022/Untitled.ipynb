```python
from datetime import datetime
def unpack_new_features(df, col_df, col):
    print("Running unpack_new_features [" + col + "] at "  + str(datetime.now())) 
    col_df = col_df[col].apply(lambda x: pd.Series(x, dtype='object')).dropna(axis=1, how='all')
    
    dataframes = []
    for j in range(100):
        df_aux = col_df.groupby(col_df.index).nth(j)
        #df_aux.columns = df_aux.columns   '_'   str(j)
        df_aux.columns = str(col) + '__' + df_aux.columns + '__' + str(j)
        dataframes.append(df_aux)
        if dataframes[-1].shape[0] == 0:
            del dataframes[-1]
            break
            
    new_features_df = pd.concat(dataframes, axis=1, ignore_index=False)
    
    return df.join(new_features_df, how='inner').drop(columns=[col])

def tabular(dict_list):
    df = pd.DataFrame(dict_list)
    
    for i in range(100):
        changes = 0
        print('Loop: ', i)
        #display(df.head())
    
        for col in df.columns:
            #print('\n', col)
            
            if df[col].dropna().shape[0] == 0:
                continue
            
            col_type = df[col].dropna().apply(type).drop_duplicates()
            data_types = df[col].dropna().apply(lambda x: type(x[0]) if (isinstance(x, list) and len(x) > 0) else None).dropna()
            data_types = data_types.drop_duplicates()
            
            #print('Col Type (Shape): ', col_type.shape)
            #print('Col Type (Iloc): ', col_type.iloc[:])
            #print('Data Types (Shape): ', data_types.shape)
            #print('Data Types (Iloc): ', data_types.iloc[:])
            
            if (
                    (col_type.shape[0] == 1) and 
                    (col_type.iloc[0] == list) and
                    (data_types.shape[0] == 1) and 
                    (data_types.iloc[0] == dict)
                ):
                
                col_df = df.loc[:, [col]].explode(col)

                if df[col].dropna().apply(len).sum() > 0: # check is the dicts aren't empty
                    changes =1

                    df = unpack_new_features(df, col_df, col)
                    print("Running tabular - 1 df.shape ["  + str(df.shape)  + "] at "  + str(datetime.now()))
                    #print
            
            elif (
                    (col_type.shape[0] == 1) and 
                    (col_type.iloc[0] == dict)
                ):
                
                if df[col].dropna().apply(len).sum() > 0 :
                    changes =1
                    col_df = df.loc[:, [col]]

                    df = unpack_new_features(df, col_df, col)
                    print("tabular - 2 df.shape [" +  str(df.shape) +  "] at "  + str(datetime.now()))
                    
        if changes == 0:
            break
        
        gc.collect()

    return df.reset_index(drop=True)
```
